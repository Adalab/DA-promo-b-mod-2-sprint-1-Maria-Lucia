```python
# 2 Chequead que los nombres de las columnas que sean iguales entre dataframes son iguales. 
# ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sean todas iguales unificad el nombre de las columnas.
```

```python
# unificamos los nombres de las columnas, aplicando una función (dict comprenhension) que les quite los espacios al principio y al final del string, 
# que sustituya los espacios por _, y que convierta los nombres a minusculas

df_columnas_nuevas = {col: col.strip().replace(" ","_").lower() for col in df_historico.columns}
df_historico.rename(columns = df_columnas_nuevas, inplace = True)
df_historico.columns
```

```python
# le paso la correcion también a df_ataques para igualas condiciones de los nombres de las columnas
df_columnas_nuevas = {col: col.strip().replace(" ","_").lower() for col in df_ataques}
df_ataques.rename(columns = df_columnas_nuevas, inplace = True)
df_ataques.columns
```

```python
# 3. En este momento nos habremos dado cuenta que el DataFrame que generamos en el pair programming Pandas IV es similar al DataFrame de los datos historicos. 
# Si recordamos, eliminamos algunas columnas en la sesión III que no eran útiles, pero esas columnas están todavía en el histórico. 
# Si no fuera por eso podríamos unir esos dataframes. Realizad los cambios que creais necesarios para que los dataframes sean iguales y luego juntadlos usando el mejor método
```

```python
#Hago una copia antes de proceder a borrar info en df_historico

df_historico2 = df_historico.copy()
df_historico2.head()
```

```python
# Sacamos la lista de columnas que queremos borrar de historico2 (que son las columnas que NO estan en tiburones)

b = df_tiburones.columns
a = df_historico.columns

c = [x for x in a if x not in b]

c
```

```python
# Borro las columnas con la lista que me he creado

df_historico2.drop(columns = c, inplace=True)
df_historico.columns
```

```python
# Genero df_general. Decido quedarme sólo con las columnas de df_tiburones, porque sino, me voy a llenar el dataset de nan, y no les veo mucha relevancia a las columnas "extra" de df_historico

df_general = pd.concat([df_tiburones, df_historico2], axis=0, ignore_index = True) # uno los dos dataframes con concat, pues ya tienen el mismo numero de columnas
df_general.drop(columns = "unnamed:_0", inplace = True) # nos cargamos el unnamed, porque le vamos a resetear el index, no nos sirve.
df_general.reset_index() # reseteamos el indice y así tenemos un índice común a todo el data frame
```

```python
#  4. Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el DataFrame de información adicional. 
# Debatid que método es el más correcto para unir toda esta información y ejecutadlo
```

```python
#Hago una copia para trbajar con la copia
df_ataques2 = df_ataques.copy()
```

```python
# Vemos que hay 4 rows más en ataques que en general, vemos si podemos arreglar eso antes de mergearlos. Para ahorrarnos muchos nan en nuestro data set
df_ataques2.info()
```

```python
df_ataques2 = df_ataques2.dropna(subset=["date", "case_number"])
```

```python
# Despues de reducir la diferencia entre los Dataframes, procedemos a unirlos, con un merge
df_ataques2.info()
```

```python
info_tiburones = df_ataques2.merge(df_general, how = "left", on= "case_number")
info_tiburones.shape
```
