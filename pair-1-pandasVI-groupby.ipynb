```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./ficheros_pandas/info_tiburones.csv")
df.info()
```

```python
# Ejercicio 1. Cuántos ataques hubo por país? Devuelve un DataFrame. (ordénalos de mayor a menor. Usa .sort_values())

ataques = pd.DataFrame(df.groupby("country")["country"].count().sort_values(ascending=False))
ataques.head()
```

```python
# Ejercicio 2. ¿Cuántos ataques ha habido por año? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.
ataques_ano = pd.DataFrame(df.groupby("year")["year"].count().sort_values(ascending=False))
ataques_ano.head()
```

```python
# ¿Cuántos ataques hubo por año y sexo? Devolved un DataFrame solo del año 2000 en adelante.
ataque_2000 = df[df["year"] >= 2000]
ataque_2000
```

```python

ataques_ano_sex = ataque_2000.groupby(["year","sex_"])['type'].count()
df_ataques_ano_sex = pd.DataFrame(ataques_ano_sex)
df_ataques_ano_sex.head()
```

```python
# Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. 
# ¿Cuántos de los ataques fueron letales por sexo? Devuelve un DataFrame.
```

```python
ataques_uas = df[(df["country"] == "USA") | (df["country"] == "SOUTH AFRICA") | (df["country"] == "AUSTRALIA")]
ataques_uas
```

```python
ataques_uas_groupby = pd.DataFrame(ataques_uas.groupby(["sex_", "Fatal (Y/N)", "country"])["case_number"].count()).reset_index()

letales = ataques_uas_groupby[ataques_uas_groupby["Fatal (Y/N)"] == "Y"]
letales
```
