```python
import pandas as pd
pd.options.display.max_columns = None
import numpy as np
import re
import random as random
```

```python
df = pd.read_csv("./ficheros_pandas/info_tiburones.csv")
df.head(3)
```

```python
# Ejercicio 1. Cread una columna nueva y una funciÃ³n que nos de el mes donde ocurriÃ³ el ataque. 
# Tened en cuenta que no todas las filas tienen la misma estructura y que puede que no haya la informaciÃ³n de mes. 
# En ese caso devolved un nulo (NaN).
```

```python
patron = "[\d]-[A-Z].*[a-z]-[\d]"
patron2 = "[A-Z].*[a-z]"
```

```python
lista= []
```

```python
len(lista)
```

```python
for i in df["Date"]:    
    lista.append(re.findall(patron, i))

```

```python
len(lista)
```

```python
lista2 = []

for i in range(0, len(lista)):
    if len(lista[i]) > 0:
        lista2.append(re.findall(patron2, lista[i][0])[0])
    else: 
        lista2.append("Nan")
        
df["mes"] = lista2
```

```python
def cambiar_mes(x):
    patron = "[\d]-[A-Z].*[a-z]-[\d]"
    patron2 = "[A-Z].*[a-z]"
    re.findall(patron, x)
    if len(re.findall(patron, x)) > 0:
        return re.findall(patron2, x)[0]
    else: 
        return np.nan
```

```python
df["mes"] =df["Date"].apply(cambiar_mes)
```

```python
# Ejercicio 2. Vamos con la columna fatal_(y/n). Lo primero que tenÃ©is que hacer es evaluar los valores Ãºnicos que hay. 
# EsperarÃ­amos tener solo dos valores: y y n, pero tristemente no deberÃ­a ser asÃ­ ðŸ˜­. 
# Cread una columna nueva y una funciÃ³n que devuelva Ãºnicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor
```

```python
df["Fatal (Y/N)"].unique()
```

```python
def limpiar(x):
    if type(x) == str:
        limpio = x.strip().lower()
        if limpio == "y":
            return "y"
        elif limpio == "n":
            return "n"
        else:
            return np.nan            
    else:
        return np.nan
```

```python
df["muerte"] =df["Fatal (Y/N)"].apply(limpiar)
```

```python
# Ejercicio 3. Seguid la misma lÃ³gica que en el ejercicio anterior pero con la columna sex. 
# Pero en este caso vamos a incluir un retito mÃ¡s: En este caso querremos tener dos valores Ãºnicos F y M. 
# De nuevo en la columna hay otros carÃ¡cteres raros como 'M ', 'lli', 'N', '.'. 
# En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendrÃ©is que usar la librerÃ­a random). 
# En caso de que sea nulo, que la funciÃ³n devuelva nulo (np.nan)
```

```python
df["sex_"].unique()
```

```python
def sexar(x):
    aleatorio = random.randint(0,1)
    
    if aleatorio == 0:
        valor = "f"
    else:
        valor = "m"
    
    if type(x) == str:
        limpio = x.strip().lower()
        if limpio == "f":
            return "f"
        elif limpio == "m":
            return "m"
        else:
            return valor  
    else:
        return valor
```

```python
df["sexo"] =df["sex_"].apply(sexar)
```
