```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
info_tiburones = pd.read_csv("./ficheros_pandas/info_tiburones.csv")
```

```python
df_2015 = info_tiburones[info_tiburones["year"] >= 2015]
```

```python
# Ejercicio 1. Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. 
# Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos
```

```python
df = pd.DataFrame(df_2015.groupby(["activity", "year"])["case_number"].count()).reset_index()
df.head()
```

```python
df_2015.groupby(["activity", "year"])["href"].size().reset_index() #es lo mismo que lo de arriba pero usando size qeu vale para objetos "groupby"
```

```python
plt.figure(figsize =(20,4))
fig = sns.barplot(x='year' , y='case_number' , data = df, hue ="activity")
fig.set(xlabel='Años', ylabel = 'Número de casos')
fig.set_title('¿que hacías mientras te atacaba un tiburón?')

plt.show()
```

```python
# Ejercicio 2 Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. 
# Hacedlo solo desde el 2015.
```

```python
df4 = info_tiburones[(info_tiburones["year"] >= 2015) & ((info_tiburones["sex_"] == "M") | (info_tiburones["sex_"] == "F"))]
df4.head()
```

```python
df4["year"].value_counts()
```

```python
plt.figure(figsize =(8,4))
fig = sns.countplot(x='year', data = df4, hue = 'sex_')
fig.set(xlabel='Años', ylabel= 'Número de ataques' )
fig.set_title('Número de ataques en función del sexo')
plt.show()
```

```python
# Ejercicio 3 ¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.
```

```python
df5 = info_tiburones[info_tiburones["year"] >= 1500 ]
df5.head(3)
```

```python
plt.figure(figsize =(100,8))
fig = sns.countplot(x='year', data = df5)
fig.set(xlabel='Años', ylabel= 'Número de ataques' )
fig.set_title('Número de ataques a lo largo del a historia')
plt.show()
```

```python
# Ejercicio 4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. 
# Muestra el número de ataques por país y año desde el año 1900.
```

```python
ataques_uas = info_tiburones[((info_tiburones["country"] == "USA") | (info_tiburones["country"] == "SOUTH AFRICA") | (info_tiburones["country"] == "AUSTRALIA"))  & (info_tiburones["year"]>1900)]
    
ataques_uas2 = pd.DataFrame(ataques_uas.groupby(["country","year"])["case_number"].count().reset_index())
ataques_uas2.sample(20)
```

```python
plt.figure(figsize =(15,4))
fig = sns.scatterplot(x='year' , y='case_number' , data = ataques_uas2, hue= 'country')
fig.set(xlabel='Años', ylabel = 'Número de ataques')
fig.set_title('Gráfica de disperisón años-nº de ataques por país')
plt.show()
```
